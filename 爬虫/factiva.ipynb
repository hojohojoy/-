```python
#适用于factiva：得到页面上新闻的链接
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup

#通过主页面链接得到新闻链接
def getNewsPage(url):
    try:
        #得到html
        f = open('html.txt')
        html = f.read()
        f.close()

        soup = BeautifulSoup(html,'html.parser')
        #print(soup.prettify())
    
        #找到链接所在位置
        headlines = soup.find_all('tr',{"class":{"headline"}})
        #print(headlines)

        #提取出链接
        for headline in headlines:
            tags = headline('a')
            for tag in tags:
                link = tag.get('href') #得到链接
                #排除无关链接，并补齐链接
                if link[0] != 'j':
                    link='https://global.factiva.com/'+link[3:]
                    print(link)
        
    except:
        print('wrong, try again')
        #getNewsPage(url)

def main():
    url = 'https://global.factiva.com/ha/default.aspx?page_driver=searchBuilder_Search#./!?&_suid=167870324547607607899761907566'
    getNewsPage(url)

main()
```
