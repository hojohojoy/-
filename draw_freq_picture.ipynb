```python
from matplotlib import pyplot as plt

plt.rcParams ['axes.unicode_minus'] = False
plt.rcParams['font.sans-serif']=['SimHei']
plt.style.use('ggplot')

x=[0,1]
y1=[0.70384,0.65577]
y2=[0.29626,0.34423]
plt.figure(figsize=(2,5),dpi=1000) #设置图片大小和像素
plt.plot(x,y1,label='积极词频率',color='green') #设置折线颜色和标签
plt.plot(x,y2,label='消极词频率',color='blue')
plt.xticks(range(2),labels=['疫情前','疫情后']) #x轴坐标轴
plt.ylim((0, 1))#y轴坐标轴
plt.ylabel('频率')
plt.legend(fontsize=4,loc='center right')#添加图例
plt.text(0,0.70384,0.70384,ha='left',va='bottom',fontsize=4)
plt.text(1,0.65577,0.65577,ha='left',va='bottom',fontsize=4)
plt.text(0,0.29626,0.29626,ha='left',va='bottom',fontsize=4)
plt.text(1,0.34423,0.34423,ha='left',va='bottom',fontsize=4)
plt.show()
```
